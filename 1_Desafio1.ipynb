### 14. Escribí una función que use solo lists comprehensions (no for loops) para calcular el Desvío Estándar. Imprimí el Desvío Estándar de cada columna numérica.

```python
from math import sqrt


def DesviacionEstandar(values):
    valorPromedio = sum(values) / float(len(values))
    variacion = sum([(value-valorPromedio)**2 for value in values]) / len(values)
    return sqrt(variacion)

for col in range(1,len(headers)):
    print 'Desviacion Estandar de ',headers[col] ,'\t', DesviacionEstandar(dicScores[headers[col]])
```

### 16. Graficá la distribución de la variable Math

```python
def distribution_plotter(column, data):
    sns.set(rc={"figure.figsize": (10, 7)})
    sns.set_style("white")
    dist = sns.distplot(data, hist_kws={'alpha':0.2}, kde_kws={'linewidth':5})
    dist.set_title('Distribucion de ' + column + '\n', fontsize=16)
```

```python
distribution_plotter('Math', dicScores['Math'])
```
